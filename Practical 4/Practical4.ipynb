```python
import pandas as pd

data = {
    "Name": ["Shah", "Vats", "Vats", "Kumar", "Vats","Kumar", "Shah", "Shah", "Kumar", "Vats"],
    "Gender": ["Male", "Male", "Female", "Female", "Female","Male", "Male", "Female", "Female", "Male"],
    "MonthlyIncome": [114000, 65000, 43150, 69500, 155000,103000, 55000, 112400, 81030, 71900]
}

df = pd.DataFrame(data)

print(df)
```

```python
family_income = df.groupby("Name")["MonthlyIncome"].sum()

print("\nFamily-wise Gross Monthly Income:\n")
print(family_income)
```

```python
income_stats = df.groupby("Name")["MonthlyIncome"].agg(["max", "min"])

print("\nHighest and Lowest Income for Each Family:\n")
print(income_stats)
```

```python
low_income = df[df["MonthlyIncome"] < 80000]

print("\nMembers earning less than 80000:\n")
print(low_income)
```

```python
female_data = df[df["Gender"] == "Female"]

female_count = female_data.shape[0]

print("\nTotal Females:", female_count)
print("Average Female Income:", female_data["MonthlyIncome"].mean())
```

```python
overall_avg = df["MonthlyIncome"].mean()
overall_avg

df = df[df["MonthlyIncome"] >= overall_avg]

print("\nData after deleting rows with income less than overall average:\n")
print(df)
```
