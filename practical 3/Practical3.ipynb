```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
print(df.head())
```

```python
features = df.select_dtypes(include='number')

# Mean
print("Mean:\n", features.mean())

# Median
print("\nMedian:\n", features.median())

# Mode
print("\nMode:\n", features.mode().iloc[0])

# Standard Deviation
print("\nStandard Deviation:\n", features.std())

# Standard Error
print("\nStandard Error:\n", features.sem())
```

```python
features = df.select_dtypes(include='number')

n = len(features)

for col in features.columns:
    mean = features[col].mean()
    sd = features[col].std()
    se = sd / np.sqrt(n)
    
    lower = mean - 1.96 * se
    upper = mean + 1.96 * se
    print(f"Mean for {col}: {mean}")
    print(f"Standard for {col}: {sd}")
    print(f"95% CI for {col}: ({lower:.3f}, {upper:.3f})\n")
```

```python
corr = features.corr()

print("\nCorrelation Matrix:\n")

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
cov_value = features['sepal_length'].cov(features['petal_length'])

print("\nCovariance between Sepal Length and Petal Length:", cov_value)
```
