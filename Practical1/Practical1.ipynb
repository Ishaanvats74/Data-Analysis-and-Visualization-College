```python
import numpy as np
```

# (a) Mean, Standard Deviation, Variance along 2nd axis

```python
arr_a = np.random.randint(1, 50, size=(4, 5))
print("2D Random Array:\n", arr_a)

mean = np.mean(arr_a, axis=1)
std_dev = np.std(arr_a, axis=1)
variance = np.var(arr_a, axis=1)

print("\nMean along 2nd axis:", mean)
print("Standard Deviation along 2nd axis:", std_dev)
print("Variance along 2nd axis:", variance)
```

# (b) Create m x n array, print properties, reshape to n x m

```python
m = int(input("\nEnter number of rows (m): "))
n = int(input("Enter number of columns (n): "))

arr_b = np.random.randint(1, 100, size=(m, n))

print("\nOriginal Array:\n", arr_b)
print("Shape:", arr_b.shape)
print("Type:", type(arr_b))
print("Data Type:", arr_b.dtype)

reshaped_arr = arr_b.reshape(n, m)
print("\nReshaped Array (n x m):\n", reshaped_arr)
```

# (c) Check zero, non-zero and NaN values in 1D array

```python
arr_c = np.array([0, 3, np.nan, 5, 0, 8, np.nan,10])
reshaped_arr = arr_c.reshape(2, 4)
print("\n1D Array:", arr_c)
print(reshaped_arr)

zero_indices = np.where(arr_c == 0)
nonzero_indices = np.where(arr_c != 0)
nan_indices = np.where(np.isnan(arr_c))

print("Indices of Zero elements:", zero_indices[0])
print("Indices of Non-Zero elements:", nonzero_indices[0])
print("Indices of NaN elements:", nan_indices[0])
```

# (d) Covariance and Correlation

```python
Array1 = np.random.randint(10)
Array2 = np.random.randint(10)
Array3 = np.random.randint(10)

Array4 = Array3 - Array2

Array5 = 2 * Array1

print("\nArray1:", Array1)
print("Array4 (Array3 - Array2):", Array4)
print("Array5 (2 * Array1):", Array5)

covariance = np.cov(Array1, Array4)
print("\nCovariance of Array1 and Array4:\n", covariance)

correlation = np.corrcoef(Array1, Array5)
print("\nCorrelation of Array1 and Array5:\n", correlation)
```

# (e) Sum of first half & product of second half

```python
Array1 = np.random.randint(1, 10, 10)
Array2 = np.random.randint(1, 10, 10)

print("\nArray1:", Array1)
print("Array2:", Array2)

first_half_sum = np.sum(Array1[:5]) + np.sum(Array2[:5])

second_half_product = np.prod(Array1[5:]) * np.prod(Array2[5:])

print("Sum of first half of both arrays:", first_half_sum)
print("Product of second half of both arrays:", second_half_product)
```
