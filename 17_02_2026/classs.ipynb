```python
import pandas as pd 
import numpy as np
```

```python
obj4 = pd.DataFrame(np.arange(9).reshape(3,3),index=['a','b','c'],columns=['ohio','texas','california'])
print(obj4)
obj5 = obj4.reindex(['a','b','c','d'])
print(obj5)
```

```python
obj6 = obj5.reindex(columns=['texas','california','utah','ohio'])
print(obj6)
```

```python
obj6 = obj5.reindex(columns=['texas','california','utah','ohio'],fill_value=1)
print(obj6)
```

```python
obj6 = obj5.reindex(columns=['texas','california','utah','ohio'],fill_value=1)
print(obj6)

obj8 = obj5.drop('ohio',axis=1)
print(obj8)

obj9 = obj5.drop(columns=['ohio'])
print(obj9)
```

```python
s2 = pd.Series([0,1,2],index=[10,20,30])
print(s2)
```

```python
s3 = s2.reindex([11,21,34])
print(s3)
```

```python
s4 = s2.reindex([11,21,34],method='nearest',tolerance=2)
print(s4)
```

```python
index = pd.MultiIndex.from_product([['A','B'],[1,2]])
print(index)
```

```python
multindex = pd.MultiIndex.from_product([['A','B','C'],[1,2]])
print(multindex)
```

```python
s = pd.Series([10,20,30,40],index=index)
print(s)
```

```python
new_s = s.reindex(index=multindex)
print(new_s)
```
